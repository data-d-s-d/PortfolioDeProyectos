```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in from location of the data:

df = pd.read_csv(r'C:\Users\diego\OneDrive\Escritorio\Proyecto\Proyecto Alex\Project 4\movies.csv')
```

```python
df
```

```python
# For-loop to check for missing data:

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Printing data types of columns:

print(df.dtypes)
```

```python
# Droping rows with na values and changing data types for "budget" and "gross" columns:

df = df.dropna()

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Sorting by "gross" 

df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# If you want to look at the full data set:

# pd.set_option('display.max_rows', None)
```

```python
# My hypothesis is that there will be a strong correlation between the budget and the gross revenue, so..
# Scatterplot: Correlation between 'budget' and 'gross(revenue)' 

plt. scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Revenue')
plt.xlabel('Gross Revenue')
plt.ylabel('Budget')

plt.show()
```

```python
df.head()
```

```python
# It's still hard to tell if there's a correlation, so i'll make a Regression Plot using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color": "purple"}, line_kws={"color":"blue"})
```

```python
# This means there is indeed a positive correlation between budget and gross revenue,
#but we don't know exactly what the correlation is, so:

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for numeric columns')
plt.xlabel('Movie Feautres')
plt.ylabel('Movie Feautres')

plt.show()
```

```python
# Another hypothesis is that the correlation between the Company and Gross Revenue is also high,
# but 'Company' is non numeric, so i'll be "numerizing" some of the non-numeric relevant columns

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
# All set. Now i'll make a new correlation matrix with all this new values.

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for numeric columns')
plt.xlabel('Movie Feautres')
plt.ylabel('Movie Feautres')

plt.show()
```

```python
# Correlation pairs listed:

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Let's see which pairs show high correlation:

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```

```python
# Here we can see that 'votes' and 'budget' have the highest correlation with 'gross' revenue.
# Therefore, my hypothesis about the correlation between the 'company' and the gross revenue was false.  
```
